```python
from firecrawl import FirecrawlApp

import os
from dotenv import load_dotenv, find_dotenv
_ = load_dotenv(find_dotenv()) # read local .env file
firecrawl_api_key = os.getenv("FIRECRAWL_API_KEY")

app = FirecrawlApp(api_key=firecrawl_api_key)
```

```python
# Scrape a single URL:
url = "https://speechbrain.github.io/"
scraped_data = app.scrape_url(url)

scraped_data
```

```python
import pprint

pprint.pprint(scraped_data)
```

```python
# Crawl a website:
crawl_url = "https://speechbrain.github.io/"

crawl_result = app.crawl_url(crawl_url)

crawl_result
```

# <font color="gree">``Parte 2:`` Firecrawl</font>

```python
from firecrawl import FirecrawlApp
import os
from dotenv import load_dotenv, find_dotenv
_ = load_dotenv(find_dotenv()) # read local .env file
firecrawl_api_key = os.getenv("FIRECRAWL_API_KEY")

app = FirecrawlApp(api_key=firecrawl_api_key)



# Scrape a website (only):
scrape_status = app.scrape_url(
    "https://firecrawl.dev",
    params={"formats": ["markdown", "html"]}
)

scrape_status
```

```python
# Crawl a website:
crawl_status = app.crawl_url(
    "https://firecrawl.dev",
    params={
        "limit": 5,
        "scrapeOptions": {"formats": ["markdown", "html"]}
    }
)

crawl_status
```

# <font color="gree">``Parte 3:`` Firecrawl with LangChain</font>

```python
from langchain_community.document_loaders.firecrawl import FireCrawlLoader

import os
from dotenv import load_dotenv, find_dotenv
_ = load_dotenv(find_dotenv()) # read local .env file
firecrawl_api_key = os.getenv("FIRECRAWL_API_KEY")



loader = FireCrawlLoader(api_key=firecrawl_api_key,
                         url="https://firecrawl.dev",
                         mode="scrape"
                        )
```

```python
pages = []
for doc in loader.lazy_load():
    pages.append(doc)
    if len(pages) >= 10:
        # do some paged operation, e.g.
        # index.upsert(page)

        pages = []
```
